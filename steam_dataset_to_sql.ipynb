# Connect to MySQL

```python
def get_sql_connection():
  credentials_file = "credentials.json" 

  with open(credentials_file, "r") as f: 
      credentials = json.load(f) 
      sql_credential = credentials.get("sql")
      database = sql_credential.get("database")
      password = sql_credential.get("password") 


  mydb = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = password, 
    database = database
  )  

  if mydb: 
     print("Connection to database success")
     
  return mydb   
   
def check_sql_connection(db_connection):
    # If connection is open and alive 
    # https://stackoverflow.com/questions/33239480/how-to-check-if-a-mysql-connection-is-open-in-python
    if db_connection.is_connected():
        return db_connection 
    
    # else reconnect to database
    else:
        print("Connection closed. Reconnecting to database")
        return get_sql_connection()
```

```python
mydb = get_sql_connection() 
cursor = mydb.cursor()
```

# Create Database Table

Column Information  
1. **AppID**: unique identifier for the game  
2. **Name**: the name of the game  
3. **Release date**: when the game was released  
4. **Estimated owners**: estimated owners who own the game  
5. **Peak CCU**: number of concurrent users, yesterday  
6. **Required age**: age required to play (0 if for all audiences)  
7. **Price**: price of game in USD (0.0 if free)  
8. **DiscountDLC count**: number of DLC the game has (0 if none)  
9. **About the game**: a detailed description of the game  
10. **Supported languages**: comma-separated enumeration of supporting languages  
11. **Full audio languages**: comma-separated enumeration of languages with audio support  
12. **Reviews**: users’ reviews about the game  
13. **Header image**: Header image URL in the Steam store  
14. **Website**: game website  
15. **Support URL**: game support team URL  
16. **Support email**: game support team email  
17. **Windows**: if the game is supported on Windows  
18. **Mac**: if the game is supported on Mac  
19. **Linux**: if the game is supported on Linux  
20. **Metacritic score**: weighted average of reviews from professional critics and publications (the higher the better, plus you don’t need to own the game to score)  
21. **Metacritic URL**: URL that aggregates reviews for the game, determining the Metacritic score (if no score, empty)  
22. **User score**: ratio of positive to negative reviews and usually categorized in semantic buckets like ‘Overwhelmingly positive’, ‘Very positive’, and such (0 if there are no scores for the game, and is according to the SteamSpy website)  
23. **Positive**: positive votes that users give to the game (0 if no information, and comes from SteamSpy)  
24. **Negative**: negative votes that users give to the game (0 if no information, and comes from SteamSpy)  
25. **Score rank**: score rank of the game based on user reviews  
26. **Achievements**: number of achievements in the game (0 if none)  
27. **Recommendations**: users’ recommendations about the game (0 if none)  
28. **Notes**: extra information about the game content  
29. **Average playtime forever**: average playtime of the player base who owns the games in minutes  
30. **Average playtime two weeks**: average playtime of the player base who owns the games in minutes during the last two weeks  
31. **Median playtime forever**: median playtime of the player base who owns the games in minutes  
32. **Median playtime two weeks**: median playtime of the player base who owns the games in minutes during the last two weeks  
33. **Developers**: the company that developed the game  
34. **Publishers**: the company that published the game  
35. **Categories**: list of categories that the game belongs to  
36. **Genres**: list of genres that the game belongs to  
37. **Tags**: labels that help users and developers categorize and discover games on the Steam platform  
38. **Screenshots**: screenshot/snippets of the game  
39. **Movies**: videos of what the game is about  

## Create Core Tables

```python
# https://dev.mysql.com/doc/connector-python/en/connector-python-example-ddl.html  
core_table = {} 

core_table['games'] = """
CREATE TABLE games ( 
    game_id INT AUTO_INCREMENT PRIMARY KEY, 
    app_id INT UNIQUE,
    name TEXT, 
    date_released DATE,  
    required_age INT, 
    price DECIMAL(10, 2), 
    description TEXT, 
    achievements INT
);
"""

core_table['game_extra_info'] = """ 
CREATE TABLE game_extra_info ( 
    app_id INT PRIMARY KEY, 
    estimated_owners VARCHAR(50),   
    extra_description TEXT, 
    avg_playtime_mins INT,  
    header_image TEXT, 
    website TEXT,  
    support_url TEXT, 
    support_email TEXT, 
    window_supported BOOLEAN, 
    mac_supported BOOLEAN, 
    linux_supported BOOLEAN
); 
""" 

core_table['game_review'] = """ 
CREATE TABLE game_review (  
    app_id INT PRIMARY KEY, 
    reviews TEXT, 
    metacritic_score INT, 
    metacritic_url TEXT, 
    positive_vote INT, 
    negative_vote INT, 
    recommendations INT
);
""" 

core_table['companies'] = """ 
CREATE TABLE companies ( 
    company_id INT AUTO_INCREMENT PRIMARY KEY,  
    name VARCHAR(255) UNIQUE
);
"""  

core_table['languages'] = """ 
CREATE TABLE languages ( 
    language_id INT AUTO_INCREMENT PRIMARY KEY, 
    name VARCHAR(255) UNIQUE
);
""" 

core_table['categories'] = """ 
CREATE TABLE categories ( 
    category_id INT AUTO_INCREMENT PRIMARY KEY, 
    name VARCHAR(255) UNIQUE
); 
"""  

core_table['genres'] = """ 
CREATE TABLE genres ( 
    genre_id INT AUTO_INCREMENT PRIMARY KEY, 
    name VARCHAR(255) UNIQUE
);
""" 

core_table['tags'] = """ 
CREATE TABLE tags ( 
    tag_id INT AUTO_INCREMENT PRIMARY KEY, 
    name VARCHAR(100) UNIQUE
);
"""  

core_table['screenshots'] = """ 
CREATE TABLE IF NOT EXISTS game_screenshots ( 
    screenshot_id INT AUTO_INCREMENT PRIMARY KEY,
    app_id INT,
    screenshot_url VARCHAR(500),
    UNIQUE KEY unique_app_screenshot (app_id, screenshot_url)
);
""" 

core_table['movies'] = """ 
CREATE TABLE IF NOT EXISTS game_movies ( 
    movie_id INT AUTO_INCREMENT PRIMARY KEY,
    app_id INT,
    movie_url VARCHAR(500),
    UNIQUE KEY unique_app_movie (app_id, movie_url)
);
""" 
```

```python
for table_name, table_description in core_table.items():
    try:
        print(f"Creating table {table_name}")
        cursor.execute(table_description) 
        print("Created successfully \n")
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
            print("Already exists \n")
        else:
            print(err.msg, "\n")
```

## Create Relational Tables

```python
relational_table = {}  

relational_table['game_developer_publisher'] = """
CREATE TABLE game_developer_publisher (  
    app_id INT, 
    developer_id INT, 
    publisher_id INT,
    PRIMARY KEY (app_id),
    FOREIGN KEY (developer_id) REFERENCES companies(company_id),
    FOREIGN KEY (publisher_id) REFERENCES companies(company_id)
);
""" 

relational_table['game_languages'] = """
CREATE TABLE game_languages (  
    app_id INT, 
    supported_language_id INT,
    audio_language_id INT,
    FOREIGN KEY (supported_language_id) REFERENCES languages(language_id),
    FOREIGN KEY (audio_language_id) REFERENCES languages(language_id)
);
"""  

relational_table['game_categories'] = """
CREATE TABLE game_categories (  
    app_id INT, 
    category_id INT,
    PRIMARY KEY (app_id, category_id),
    FOREIGN KEY (category_id) REFERENCES categories(category_id)
);
"""   

relational_table['game_genres'] = """
CREATE TABLE game_genres (  
    app_id INT, 
    genre_id INT,
    PRIMARY KEY (app_id, genre_id),
    FOREIGN KEY (genre_id) REFERENCES genres(genre_id)
);
"""   

relational_table['game_tags'] = """
CREATE TABLE IF NOT EXISTS game_tags (  
    app_id INT, 
    tag_id INT,
    PRIMARY KEY (app_id, tag_id),
    FOREIGN KEY (tag_id) REFERENCES tags(tag_id)
);
""" 
```

```python
for table_name, table_description in relational_table.items():
    try:
        print(f"Creating table {table_name}")
        cursor.execute(table_description) 
        print("Created successfully \n")
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
            print("Already exists \n")
        else:
            print(err.msg, "\n")
```

## Core Table

In each of the core table, I converted them into an array of tuples so that I can perform batch insert into MySQL to speed up the process

```python
def insert_all_data_verification (mydb, table_name, data): 
    try:  
        mydb = check_sql_connection(mydb)  
        cursor = mydb.cursor() 
        
        select_query = f"SELECT COUNT(*) FROM {table_name}"  
        cursor.execute(select_query)  
        result = cursor.fetchone()
        data_count = result[0] 

        if (data_count == len(data)): 
            print(f"All data was inserted successfully: {data_count} out of {len(data)}") 
        else: 
            print(f"Not all data was inserted successfully: {data_count} out of {len(data)}") 
    
    except Exception as e: 
        print(f"Error has occured: {e}")
```

### Company

Normalize string to help with duplicate entry issue as it detecting that they are the same name despite being visually or unicode wise different

```python
def normalize_string(s):
    if pd.isna(s): return s
    s = str(s).strip()
    s = unicodedata.normalize('NFKD', s)
    return ''.join(c for c in s if not unicodedata.combining(c))

# Apply normalization to both columns
steam_dataset['Developers'] = steam_dataset['Developers'].apply(normalize_string)
steam_dataset['Publishers'] = steam_dataset['Publishers'].apply(normalize_string)
```
